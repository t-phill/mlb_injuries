```python
import pandas as pd
import numpy as np
import pickle

from pybaseball import playerid_lookup
```

```python
DLdf2018 = pd.read_csv('dl_list_2018.csv')
DLdf2018['Year'] = 2018
DLdf2018['Status'] = DLdf2018.Status.str[9:]
```

```python
DLdf2017 = pd.read_csv('dl_list_2017.csv')
DLdf2017['Year'] = 2017
DLdf2017['Status'] = DLdf2017.Status.str[9:]
```

```python
DLdf2016 = pd.read_csv('dl_list_2016.csv')
DLdf2016['Year'] = 2016
DLdf2016['Status'] = DLdf2016.Status.str[9:]
```

```python
DLdf2015 = pd.read_csv('dl_list_2015.csv')
DLdf2015['Year'] = 2015
DLdf2015['Status'] = DLdf2015.Status.str[9:]
```

```python
df1 = pd.merge(DLdf2018, DLdf2017, how='outer')
df2 = pd.merge(df1, DLdf2016, how='outer')
DL_df = pd.merge(df2, DLdf2015, how='outer')
DL_df.head()
```

```python
DL_df['List'], DL_df['Status'] = DL_df['Status'].str.split('|', 1).str
DL_df['Dates'], DL_df['Status'] = DL_df['Status'].str.split('|', 1).str
DL_df['StartDate'], DL_df['EndDate'] = DL_df['Dates'].str.split('-', 1).str

DL_df['Dates'] = DL_df['Dates'].str.replace('..-day', '')

DL_df['MoneyEarned'] = DL_df['Money_Earned'].str[1:]
DL_df['MoneyEarned'] = DL_df['MoneyEarned'].str.replace(',', '')

del DL_df['Dates']
del DL_df['Status']
del DL_df['Money_Earned']
```

```python
DL_df = DL_df[DL_df['List'].str.contains('day')]
DL_df['List'] = DL_df['List'].str[:-5]
```

```python
DL_df.to_pickle("./DLdf.pkl")
```
