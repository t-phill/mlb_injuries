```python
import re
def correction2(string):
    corstr = re.sub('\ +',' ',string)
    final = re.sub('\.','. ',corstr)
    return final
```

```python
def jr_replace(x):
    match = re.sub(r'jr.$',"",x)
    return match

def title_replace(x):
    match = re.sub(r'iii$',"",x)
    return match
```
