```python
import datetime

times = list()

start = datetime.datetime.strptime("2018-03-29", "%Y-%m-%d")
end = datetime.datetime.strptime("2018-10-29", "%Y-%m-%d")
date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days)]

```

```python
for date in date_generated:
    date = str(date.strftime("%Y-%m-%d"))
    times.append(date)
```

```python
times
```

```python
date_generated
```

```python
frames = pd.DataFrame()

for time in times:
    try:
        print(time)
        frame = pitching_stats_range(time, time)
        frame['date'] = time
        frames = frames.append(frame)
    except:
        print('errorrrrr: ', time)
        continue
```
