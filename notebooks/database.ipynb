```python
def jr_replace(x):
    match = re.sub(r'jr$',"",x)
    return match

def title_replace(x):
    match = re.sub(r'iii$',"",x)
    return match
    
games_18['name'] = games_18['name'].apply(jr_replace)
games_18['name'] = games_18['name'].apply(title_replace)
games_18
```
